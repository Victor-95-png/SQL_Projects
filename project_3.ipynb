```python
#BMI
#Body mass index - weight/height

def calc_bmi():
    height = df['Height']
    weight = df['Weight']
    bmi = []
    for h, w in zip(height, weight):
        bmi_1 = w/(h*h)
        bmi.append(bmi_1)
    return bmi_1

```

```python
bmi_2 = calc_bmi()
```

```python
df['Bmi']= bmi_2
```

```python
def year_calc():
    born_year = ['1998']
    year1 = []
    for y in born_year:
        years = born_year*10000
        year1.append(years)
    return years
```

```python
years = year_calc()
```

```python
df['Born_year'] = years
```

```python
def curr_calc():
    born_year = ['2025']
    year1 = []
    for y in born_year:
        years = born_year*10000
        year1.append(years)
    return years
```
