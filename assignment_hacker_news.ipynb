```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the most common titles
df["title"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Most Common Titles")
plt.xlabel("Title")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
